```python
# reads original data and prints columns names with more than 50 percent full data
df = pd.read_csv("../marketing_data/utdallas_contest_data-1.csv",low_memory=False,index_col="dataright_seq")
counts_raw = df.count().as_matrix()
counts = np.true_divide(counts_raw,df.shape[0])
one_hot = counts
one_hot[one_hot<0.5] = 0

for i in range(len(one_hot)):
	if one_hot[i] !=0:
		print df.columns[i]
```

```python
df_excell = pd.ExcelFile('../marketing_data/dictionary.xlsx',encoding='utf-8')
df_excell = pd.read_excel(df_excell,'demos',index_col='Allant ID')
valid_columns = pd.read_csv('stored/more_than_fifty.txt')
'Homeowner_Renter_90303' in  valid_columns.header.values


category = {}
for index, row in df_excell.iterrows():
	if row['Allant Field Name'] in valid_columns.header.values:
		if row.Category in category:
			category[row.Category].append(row['Allant Field Name'])
		else:
			category[row.Category] = []
			category[row.Category].append(row['Allant Field Name'])
category
```

```python
history_columns_df = pd.read_csv("history_columns.csv")
comp_30_month_column_names = [ column for column in history_columns_df.this_is_a_header.values if ('_shopped_3_mos' in column and 'comp' in column)]
other_30_month_column_names = [ column for column in history_columns_df.this_is_a_header.values if ('_shopped_3_mos' in column and 'other' in column)]
all_30_month_column_names = comp_30_month_column_names + other_30_month_column_names
all_30_month_column_names
```

```python
df.columns.values
```

```python
df['sum_all_shopped_3_mons'] = 0
```

```python
for index in df.index:
    for col in all_30_month_column_names:
        df.at[index, 'sum_all_shopped_3_mons'] += df.at[index,col]
```

```python

list_columns_for_check = all_30_month_column_names+ ['sum_all_shopped_3_mons']
df[df.sum_all_shopped_3_mons > 1][list_columns_for_check]
```
